## Création du modèle

Nous allons également avoir besoin d'une métrique personnalisée **w_f1_score** qui permette de calculer le F1 Score pondéré. On veille à ajouter l'annotation **@keras.saving.register_keras_serializable** pour que Keras ne pose pas de problème lors du chargement du modèle.


```python
def f1_score_sklearn(y_true, y_pred):
    y_true_classes = np.argmax(y_true, axis=1)
    y_pred_classes = np.argmax(y_pred, axis=1)
    return f1_score(y_true_classes, y_pred_classes, average='weighted')

@keras.saving.register_keras_serializable(name='w_f1_score')
def w_f1_score(y_true, y_pred):
    f1 = tf.py_function(f1_score_sklearn, (y_true, y_pred), tf.float64)
    return f1
```

Toujours en vue d'être résilient par rapport aux interruptions qui ont lieu régulièrement lors de la phase d'entraînement du modèle, créons des fonctions qui vont nous permettre de sauvegarder et restaurer un modèle Keras, afin de pouvoir reprendre l'entraînement s'il a été stoppé, ou même d'afficher les métriques d'un modèle dont l'entraînement est complet.

Les deux fonctions suivantes vont nous permettre d'enregistrer et de charger les fichiers au format pickle. Elles nous seront utiles pour ne pas recalculer inutilement les prédictions (une opération qui peut être longue) faites par nos modèles individuels.


```python
def load_pickle(path):
  path = output_dir+"/"+path
  path_obj = Path(path)
  if path_obj.exists():
    print("Load pickle file from",path)
    with open(path, 'rb') as file:
      return pickle.load(file)
  else:
    print("No pickle file found at",path)
    return None
```

```python
def save_pickle(path,arr):
  path = output_dir+"/"+path
  print("Saving pickle file to",path)
  with open(path, 'wb') as file:
    pickle.dump(arr,file)
```

Ajoutons une fonction qui va nous permettre de réordonner les colonnes de la matrice de prédiction retournée par **model.predict** pour les modèles Keras ou **model.predict_proba** pour les modèles sklearn.


```python
def reorder_predict_cols(pred,old_ordered_classes,new_ordered_classes):
  old_classes_to_index = {}
  for i in range(len(old_ordered_classes)):
    old_classes_to_index[old_ordered_classes[i]] = i

  reordered_indexes = [old_classes_to_index[c] for c in new_ordered_classes]
  reordered_pred = pred[:, reordered_indexes]
  return reordered_pred
```

Créons une fonction **get_average_pred** qui combine les prédictions du modèle d'image et du modèle de texte en calculant la prédiction moyenne des deux.


```python
def get_average_pred(img_pred,text_pred,img_pred_weight=0.5,text_pred_weight=0.5):
  combined_pred = (img_pred * img_pred_weight) + (text_pred * text_pred_weight)
  final_pred = np.argmax(combined_pred, axis=1)
  return final_pred
```

la fonction **prepare_dataset** permet de convertir la variable **text_fr** en un format accepté par le **trainer** utilisé pour effectuer les prédictions sur le modèle **CamemBERT**

```python
class Dataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}

    def __len__(self):
        return len(self.encodings.input_ids)

def prepare_dataset(text, tokenizer):
    # Tokenisation des textes de la colonne 'text_fr'
    encodings = tokenizer(text, truncation=True, padding=True, max_length=512)
    # Création du dataset
    dataset = Dataset(encodings)
    return dataset
```

**get_predictions** permet d'obtenir les prédictions du modèle **CamemBERT**.

```python
def get_predictions(trainer, dataset):
    # Obtenir les prédictions (logits)
    predictions_output = trainer.predict(dataset)
    predictions = predictions_output.predictions
    # Convertir les logits en probabilités
    probabilities = torch.softmax(torch.tensor(predictions), dim=1).numpy()
    return predictions, probabilities
```

## Entraînement du modèle

Notre classifieur final va s'entraîner sur les probabilités de prédictions en sortie de nos deux modèles texte et image. Produisons donc ces prédictions sur l'ensemble d'entraînement.

D'abord pour les images:


```python
train_pred_image_path = "train_pred_image_"+get_reporting_value(model_id,"image_model_id")+".pkl"
train_pred_image = load_pickle(train_pred_image_path)
if train_pred_image is None:
  train_pred_image = model_image.predict(train_generator,verbose=1)
  save_pickle(train_pred_image_path,train_pred_image)
```

Puis pour le texte

```python
train_pred_text_path = "train_pred_text_"+get_reporting_value(model_id,"text_model_id")+".pkl"
train_pred_text = load_pickle(train_pred_text_path)
if train_pred_text is None:
  train_pred_text = model_text.predict_proba(X_train_tfidf)
  train_pred_text = reorder_predict_cols(train_pred_text,model_text.classes_,np.array(classes_ordered,dtype="int") )
  save_pickle(train_pred_text_path,train_pred_text)
```

On peut maintenant concaténer les probabilités prédites par nos deux modèles dans une seule matrice. C'est cet ensemble de features qui va constituer l'ensemble d'entraînement de notre modèle de **LogisticRegression**.
